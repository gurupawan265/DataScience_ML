```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
##create the dataset

X,Y=make_classification(n_samples=100,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
Y_pred=logistic.predict(X_test)
Y_pred
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(Y_test,Y_pred)
print("accuracy score:",score)
cm=confusion_matrix(Y_test,Y_pred)
print("confusion_matrix:",cm)
print(classification_report(Y_test,Y_pred))
```

```python
#Hyperparameter tuning and cross validation
param_grid = {
    'solver': ['liblinear', 'saga'],
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10]
}
model=LogisticRegression()
param_grid
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
##Grid Search CV
# try to find best parameters
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=param_grid,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python

grid.fit(X_train, Y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
Y_pred=grid.predict(X_test)
Y_pred
```

```python
#after appying best parameters
score=accuracy_score(Y_test,Y_pred)
print("accuracy score:",score)
cm=confusion_matrix(Y_test,Y_pred)
print("confusion_matrix:",cm)
print(classification_report(Y_test,Y_pred))
```

Randomized Search  CV


```python
from sklearn.model_selection import RandomizedSearchCV
randomcv=RandomizedSearchCV(estimator=model,param_distributions=param_grid,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,Y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
Y_pred=randomcv.predict(X_test)
```

```python
#after appying best parameters
score=accuracy_score(Y_test,Y_pred)
print("accuracy score:",score)
cm=confusion_matrix(Y_test,Y_pred)
print("confusion_matrix:",cm)
print(classification_report(Y_test,Y_pred))
```
